```python
import os
import re
import networkx as nx
import pickle


def process_gpickle(file_path, target_path):
    filename = os.path.basename(file_path)
    print(
        "Graph ",
        file_path,
        "is being imported ...",
    )
    with open(file_path, "rb") as f:
        G = pickle.load(f)

        G = nx.convert_node_labels_to_integers(G, first_label=1, ordering="default")

        filename = f"{filename[:-8]}.clq"
        print(f"Generating graph: {filename}")
        print(target_path)
        if not os.path.exists(target_path):
            os.makedirs(target_path)
        with open(target_path + filename, "w") as f:
            # write the header
            f.write("p EDGE {} {}\n".format(G.number_of_nodes(), G.number_of_edges()))
            # now write all edges
            for u, v in G.edges():
                f.write("e {} {}\n".format(u, v))

# IF directory has sub directory
# input_graph_directory = "../graphs/gnm_random_graph_scalability"
# target_graph_directory = "../../fobc/graphs/gnm_random_graph_scalability"
input_graph_directory = "../graphs/er_graphs"
target_graph_directory = "../../fobc/graphs/er_graphs"
for root, dirs, files in os.walk(input_graph_directory):
    print(target_graph_directory, root[len(input_graph_directory) + 1 :] )
    new_target = os.path.join(
        target_graph_directory, root[len(input_graph_directory) + 1 :] + "/"
    )
    print(new_target)
    #new_target = target_graph_directory
    for filename in files:
        if filename.endswith(".gpickle"):
            process_gpickle(os.path.join(root, filename), new_target)

# ELSE
input_graph_directory = "../graphs/er_700-800"
target_graph_directory = "../../fobc/graphs/er_700-800/"
for filename in os.listdir(input_graph_directory):
    print(target_graph_directory)
    if filename.endswith(".gpickle"):
        process_gpickle(os.path.join(input_graph_directory, filename), target_graph_directory)
```
